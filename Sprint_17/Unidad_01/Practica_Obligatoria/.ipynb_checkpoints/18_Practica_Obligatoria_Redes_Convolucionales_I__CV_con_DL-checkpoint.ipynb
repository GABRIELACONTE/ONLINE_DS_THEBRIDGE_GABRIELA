```python
### CARGA DE DATOS
```

```python
# Definir la ruta base donde están las imágenes
data_dir = "data"

# Directorios de entrenamiento
train_dirs = [
    os.path.join(data_dir, "github_train_0"),
    os.path.join(data_dir, "github_train_1"),
    os.path.join(data_dir, "github_train_2"),
    os.path.join(data_dir, "github_train_3"),
]

# Directorio de test
test_dir = os.path.join(data_dir, "github_test")
```

```python
def load_images_from_folder(folder):
    images = []
    labels = []

    for filename in os.listdir(folder):
        file_path = os.path.join(folder, filename)

        # Cargar imagen y redimensionarla
        img = load_img(file_path, target_size=(32, 32))
        img_array = img_to_array(img) / 255.0  # Normalizar entre 0 y 1

        # Extraer la clase del nombre del archivo
        if filename.lower().startswith("cat"):
            label = 0  # Gato
        elif filename.lower().startswith("dog"):
            label = 1  # Perro
        else:
            continue  # En caso de un archivo raro, lo ignoramos

        images.append(img_array)
        labels.append(label)

    return np.array(images), np.array(labels)
```

```python
# Cargar datos de entrenamiento
X_train, y_train = [], []

for train_dir in train_dirs:
    X_part, y_part = load_images_from_folder(train_dir)
    X_train.append(X_part)
    y_train.append(y_part)

# Unir todas las partes
X_train = np.concatenate(X_train, axis=0)
y_train = np.concatenate(y_train, axis=0)

# Cargar datos de test
X_test, y_test = load_images_from_folder(test_dir)

# Verificar la distribución de clases de nuevo
unique, counts = np.unique(y_train, return_counts=True)
print("Distribución en Train:", dict(zip(unique, counts)))

unique, counts = np.unique(y_test, return_counts=True)
print("Distribución en Test:", dict(zip(unique, counts)))
```

```python
# Mostrar algunas imágenes de entrenamiento con su etiqueta
fig, axes = plt.subplots(3, 5, figsize=(10, 6))
fig.suptitle("Ejemplos de imágenes en el dataset", fontsize=14)

for i, ax in enumerate(axes.flat):
    idx = np.random.randint(len(X_train))  # Índice aleatorio
    ax.imshow(X_train[idx])
    ax.set_title("Perro" if y_train[idx] == 1 else "Gato")
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
### ENTRENAMIENTO DEL MODELO
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Definir el modelo
model = Sequential([

    # 🔹 PRIMER BLOQUE CONVOLUCIONAL
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    #  Conv2D(32, (3,3)): 
    #   - 32 filtros: Es un número estándar para la primera capa, suficiente para captar patrones básicos.
    #   - (3,3) tamaño del kernel: Es el tamaño más común para detectar bordes y texturas pequeñas.
    # Activación 'relu': Evita problemas de gradientes y mejora el aprendizaje.
    # input_shape=(32,32,3): 
    #   - 32x32: Tamaño de nuestras imágenes.
    #   - 3: Tres canales (RGB, imágenes en color).

    MaxPooling2D(pool_size=(2, 2)),
    # pool_size=(2,2): Reduce la dimensión de la imagen a la mitad (downsampling).
    #   - Captura las características más importantes eliminando detalles innecesarios.
    #   - Reduce la carga computacional y el riesgo de overfitting.

    # 🔹 SEGUNDO BLOQUE CONVOLUCIONAL
    Conv2D(64, (3, 3), activation='relu'),
    #  Conv2D(64, (3,3)): 
    #   - 64 filtros: Duplicamos el número de filtros para aprender características más complejas.
    #   - (3,3) tamaño del kernel: Se mantiene el mismo porque sigue siendo efectivo.
    # Activación 'relu': Continuamos con ReLU para mejorar el aprendizaje.

    MaxPooling2D(pool_size=(2, 2)),
    #  Igual que antes, reducimos la dimensionalidad a la mitad.

    # 🔹 CAPA FLATTEN (Aplanado)
    Flatten(),
    # Convierte la matriz 2D en un vector 1D para conectar con la capa densa.

    #  CAPA DENSE OCULTA (128 neuronas)
    Dense(128, activation='relu'),
    # 128 neuronas: Es un número común en redes pequeñas, suficiente para captar patrones sin sobreajustar.
    #  Activación 'relu': Permite que el modelo aprenda relaciones no lineales.

    Dropout(0.5),  
    #  Dropout del 50%: Apaga aleatoriamente la mitad de las neuronas en cada iteración.
    #   - Previene el sobreajuste.
    #   - Obliga al modelo a no depender de neuronas específicas.

    # CAPA DE SALIDA (2 clases: gato o perro)
    Dense(2, activation='softmax')
    # 2 neuronas: Una para cada clase (0 = gato, 1 = perro).
    #  Softmax: Transforma las salidas en probabilidades para clasificación multiclase.
])

# Compilar el modelo
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# Optimizer: Adam (ajusta la tasa de aprendizaje automáticamente).
# Loss: sparse_categorical_crossentropy porque tenemos etiquetas 0 y 1 en vez de one-hot encoding.
# Métrica: Accuracy para medir qué tan bien clasifica el modelo.

# Mostrar el resumen del modelo
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

# Definir el callback de EarlyStopping
early_stopping = EarlyStopping(
    monitor='val_loss',  # Monitoreamos la pérdida en validación
    patience=5,          # Si no mejora en 5 épocas, detenemos el entrenamiento
    restore_best_weights=True  # Restaurar los mejores pesos obtenidos
)
```

```python
history = model.fit(
    X_train, y_train,  # Datos de entrenamiento
    epochs=50,         # Máximo de 50 épocas
    validation_split=0.2,  # 20% de los datos se usan para validación
    batch_size=32,  # Tamaño del batch
    callbacks=[early_stopping]  # Usamos el EarlyStopping
)
```

```python
### VISUALIZACION DEL MODELO
```

```python
import matplotlib.pyplot as plt

# Graficar precisión
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.xlabel('Épocas')
plt.ylabel('Precisión')
plt.title('Evolución de la Precisión')
plt.legend()

# Graficar pérdida
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.title('Evolución de la Pérdida')
plt.legend()

plt.show()
```

```python
### CAMBIOS DE HIPERPARAMETROS PARA MEJORAR EL MODELO
```

```python
from tensorflow.keras.optimizers import Adam

# Definir el modelo
model = Sequential([
   
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    MaxPooling2D(pool_size=(2, 2)),

   
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D(pool_size=(2, 2)),

   
    Flatten(),

    #CAPA DENSE OCULTA (128 neuronas)
    Dense(128, activation='relu'),
    Dropout(0.6),  # 🔺 Subimos el dropout a 0.6 para evitar sobreajuste

    Dense(2, activation='softmax')
])

# Compilar el modelo con un learning rate más bajo
model.compile(optimizer=Adam(learning_rate=0.0005),  # 🔺 Learning rate reducido
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])


model.summary()
```

```python
early_stopping = EarlyStopping(
    monitor='val_loss',  # Monitoreamos la pérdida en validación
    patience=5,          # Si no mejora en 5 épocas, detenemos el entrenamiento
    restore_best_weights=True  # Restaurar los mejores pesos obtenidos
)
```

```python
history = model.fit(
    X_train, y_train,  # Datos de entrenamiento
    epochs=50,         # Máximo de 50 épocas
    validation_split=0.2,  # 20% de los datos se usan para validación
    batch_size=64,  # Tamaño del batch
    callbacks=[early_stopping]  # Usamos el EarlyStopping
)
```

```python
import matplotlib.pyplot as plt

# Graficar precisión
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.xlabel('Épocas')
plt.ylabel('Precisión')
plt.title('Evolución de la Precisión')
plt.legend()

# Graficar pérdida
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.title('Evolución de la Pérdida')
plt.legend()

plt.show()
```

```python
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import ReduceLROnPlateau


model = Sequential([
   
    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3), kernel_regularizer=l2(0.001)),
    MaxPooling2D(pool_size=(2, 2)),

 
    Conv2D(64, (3, 3), activation='relu', kernel_regularizer=l2(0.001)),
    MaxPooling2D(pool_size=(2, 2)),

    # CAPA FLATTEN 
    Flatten(),

    # CAPA DENSE OCULTA
    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
    Dropout(0.6),  # Dropout para reducir sobreajuste

    # 🔹 CAPA DE SALIDA
    Dense(2, activation='softmax')
])

# Compilar el modelo con un learning rate más bajo
model.compile(optimizer=Adam(learning_rate=0.0005),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Callback para reducir el learning rate si la validación no mejora
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=0.0001)

# Definir EarlyStopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Entrenar el modelo
history = model.fit(
    X_train, y_train,
    epochs=50,
    validation_split=0.2,
    batch_size=64,
    callbacks=[early_stopping, reduce_lr]  # Agregamos ReduceLROnPlateau
)

# Graficar los resultados
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.xlabel('Épocas')
plt.ylabel('Precisión')
plt.title('Evolución de la Precisión')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.title('Evolución de la Pérdida')
plt.legend()

plt.show()
```

```python
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import ReduceLROnPlateau

# Definir el modelo con más filtros y kernel_regularizer L2
model = Sequential([

    Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3), kernel_regularizer=l2(0.001)),
    MaxPooling2D(pool_size=(2, 2)),

   
    Conv2D(64, (3, 3), activation='relu', kernel_regularizer=l2(0.001)),
    MaxPooling2D(pool_size=(2, 2)),

    # TERCER BLOQUE CONVOLUCIONAL (nueva capa)
    Conv2D(128, (3, 3), activation='relu', kernel_regularizer=l2(0.001)),
    MaxPooling2D(pool_size=(2, 2)),

    # CAPA FLATTEN (Aplanado)
    Flatten(),

    # CAPA DENSE OCULTA
    Dense(128, activation='relu', kernel_regularizer=l2(0.001)),
    Dropout(0.7),  # Aumentamos el dropout a 0.7

    # CAPA DE SALIDA
    Dense(2, activation='softmax')
])

# Compilar el modelo con un learning rate aún más bajo
model.compile(optimizer=Adam(learning_rate=0.0001),  # Learning rate más bajo
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Definir EarlyStopping y ReduceLROnPlateau
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=0.0001)

# Entrenar el modelo
history = model.fit(
    X_train, y_train,
    epochs=50,
    validation_split=0.2,
    batch_size=64,
    callbacks=[early_stopping, reduce_lr]
)

# Graficar los resultados
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Entrenamiento')
plt.plot(history.history['val_accuracy'], label='Validación')
plt.xlabel('Épocas')
plt.ylabel('Precisión')
plt.title('Evolución de la Precisión')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Entrenamiento')
plt.plot(history.history['val_loss'], label='Validación')
plt.xlabel('Épocas')
plt.ylabel('Pérdida')
plt.title('Evolución de la Pérdida')
plt.legend()

plt.show()
```

```python
### EVALUACION DEL ULTIMO MODELO
```

```python
# Evaluamos el modelo en el conjunto de test
test_loss, test_accuracy = model.evaluate(X_test, y_test)
print(f"Test Loss: {test_loss:.4f}")
print(f"Test Accuracy: {test_accuracy:.4f}")

# Predecir las clases para el conjunto de test
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)  # Convertimos las probabilidades en clases

# Generar el reporte de clasificación
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred_classes))

# Generar la matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred_classes)

# Visualizar la matriz de confusión
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=["Perro", "Gato"], yticklabels=["Perro", "Gato"])
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.title('Matriz de Confusión')
plt.show()
```

El modelo tiene un buen desempeño con los perros (mayor precisión y recall en la clase perro), pero tiene algo de dificultad clasificando los gatos.

* Precision baja para gatos (0.71), pero recall más bajo (0.58). Esto sugiere que el modelo tiene tendencia a clasificar a muchos gatos como perros. Este es el mayor problema.
* El F1-score de la clase perro (0.70) es mejor que el de la clase gato (0.64), lo que refuerza la idea de que el modelo tiene un desempeño más sólido en perros.

Conclusión:

* El modelo es mejor para los perros que para los gatos. Esto se confirma en la matriz de confusión, donde hay más falsos negativos (gatos mal clasificados como perros) que falsos positivos (perros mal clasificados como gatos).
* La precisión para los gatos podría mejorarse, y es algo en lo que podríamos trabajar.

```python
### MUESTRA DE IMAGENES PEOR CLASIFICADAS
```

```python
import matplotlib.pyplot as plt
import random

# Predicciones y probabilidades para el conjunto de test
y_pred_prob = model.predict(X_test)  # Predicciones de las probabilidades (no solo la clase)
y_pred_classes = np.argmax(y_pred_prob, axis=1)  # Convertimos las probabilidades a clases

# Identificamos las imágenes mal clasificadas
incorrect_preds = np.where(y_pred_classes != y_test)[0]  # Índices de imágenes mal clasificadas

# Para cada imagen mal clasificada, obtenemos la probabilidad de la clase errónea
error_images = []
for idx in incorrect_preds:
    true_class = y_test[idx]
    predicted_class = y_pred_classes[idx]
    
    # Probabilidad de la clase incorrecta
    wrong_class_prob = y_pred_prob[idx][true_class]  # Probabilidad de la clase correcta (incorrecta)
    error_images.append((idx, wrong_class_prob, predicted_class))

# Ordenamos las imágenes por la probabilidad de la clase errónea (mayor probabilidad de error)
sorted_error_images = sorted(error_images, key=lambda x: x[1], reverse=True)

# Seleccionamos el 10% de las imágenes mal clasificadas con mayor probabilidad de error
top_10_percent = sorted_error_images[:len(sorted_error_images) // 10]

# Mostramos algunas de las imágenes mal clasificadas con su probabilidad de error
for idx, prob, pred_class in top_10_percent:
    img = X_test[idx]
    true_label = "Perro" if y_test[idx] == 0 else "Gato"
    pred_label = "Perro" if pred_class == 0 else "Gato"
    
    plt.imshow(img)
    plt.title(f"True: {true_label}, Pred: {pred_label}, Prob: {prob:.2f}")
    plt.axis('off')
    plt.show()
```
