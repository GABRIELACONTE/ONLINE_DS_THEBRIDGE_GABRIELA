### Ejercicio 2

Tenemos que entender el problema antes de seguir. Nos dicen que la compaa "QuickRespond" quiere poder generar un modelo que a partir de datos como los del dataset para poder dirigir mejor su campaa de anuncios. Hablando con ellos llegamos a la conclusin de que queremos un modelo para poder predecir si un usuario pulsar en el link de un anuncio o no. Qu tipo de problemas estamos tratando?Y de que tipo de aprendizaje?Existe una variable target que puedas identificar rapidamente?

1.	Tipo de problema:
Este es un problema de clasificacin binaria, ya que el objetivo es predecir si un usuario har clic en el enlace de un anuncio o no (es decir, 1 para hizo clic y 0 para no hizo clic).
		Las etiquetas o clases posibles son dos categoras: clic (1) o no clic (0).
		Este tipo de problema trata con datos categricos en la variable objetivo (target).

3.	Tipo de aprendizaje:
Se trata de un caso de aprendizaje supervisado, ya que tienes un dataset con ejemplos etiquetados que indican si el usuario hizo clic en el anuncio (variable target).
		El modelo aprender a relacionar las caractersticas (features) del dataset con esta variable objetivo.


4.	Variable target:
La variable target es el nmero de clics o una columna similar que indique si el usuario puls en el anuncio o no. Generalmente, estar en formato binario (1/0) o categrico (S/No). Esta variable es la que intentaremos predecir con el modelo.

### Ejercicio 7

Muestra las correlaciones de las variables continuas con un "PairPlot" y emplea como codigo de color la variable target. [Esto es otra forma de hacer un bivariante y un multivariante a la vez]. Para ello como argumento del pairplot de seaborn usa el dataframe de train (no la matriz de correlacin)

```python
# Variables continuas a utilizar (ajusta segn lo que tengas en tu DataFrame)
x_variable = "Area Income"  # Ejemplo de variable continua
y_variable = "Daily Internet Usage"   # Otra variable continua

# Crear el grfico de dispersin
plt.figure(figsize=(10,7))
plt.scatter(train_set[x_variable], train_set[y_variable], c=train_set["Clicked on Ad"], cmap="coolwarm")

# Personalizar el grfico
plt.grid(True)
plt.xlabel(x_variable)
plt.ylabel(y_variable)

# Aadir barra de color (para la variable target)
cbar = plt.colorbar(label="Clicked on Ad (0: No clic, 1: Clic)") 

# Mostrar el grfico
plt.show()
```
