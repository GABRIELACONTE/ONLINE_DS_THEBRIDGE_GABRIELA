## TRATAMIENTO DE DATOS
El tratamiento y limpieza de datos es una etapa crucial en cualquier análisis, ya que garantiza la calidad y consistencia de los datos antes de realizar cualquier exploración o inferencia. A continuación, se detalla cómo abordamos esta etapa en el conjunto de datos de Netflix, enfrentándonos a problemas comunes como valores nulos, formatos inconsistentes y redundancias en las columnas. 
Vamos a empezar por el tratamiento de los nulos respecto a su relación con la hipótesis principal del EDA:

1.	Director (30% nulos, 51% cardinalidad): Imputamos el valor de "Desconocido", para evitar errores en futuros análisis, ya que eliminar filas reduciría demasiado el dataset y mantener los nulos complicaría el procesamiento y no sabemos en que medida la vamos a usar todavia.
2.	Cast (10% nulos, 87% cardinalidad): Imputamos con “Desconocido” porque el impacto de los nulos es bajo y no afecta directamente a la hipótesis principal.
3.	Country (10% nulos, 5.84% cardinalidad): Imputamos condicionalmente por Listed_In para preservar tendencias culturales y evitar el sesgo que genera imputar con la moda general.
4.	Date_added (0.1% nulos): hemos generado una columna nueva para saber aquellos que tenian nulos y les hemos imputado la moda.
5.	Rating (0.08% nulos, 20% cardinalidad): como son pocos, los buscamos y los metemos a mano, esta columna informa de la edad que tiene que tener los espectadores para ver la película o serie, lo que nos da información sobre el público al que se dirige.

```python
# Definir una función para clasificar las categorías de RATING, como es una variable con bastante informacion para mejorar su analisis agruparemos su categorias
def classify_rating(rating):
    if rating in ['r', 'tv-ma', 'nc-17', 'nr']:
        return 'Solo para adultos'
    elif rating in ['pg-13', 'tv-14']:
        return 'Mayores de 13 años'
    elif rating in ['pg', 'tv-pg', 'tv-g', 'tv-y', 'tv-y7', 'tv-y7-fv', 'g']:
        return 'Todos los públicos'
    elif rating == 'ur':
        return 'Sin clasificar'
    else:
        return 'Sin clasificar'  # Para manejar valores nulos o inesperados

# Aplicar la función a la columna rating
df['rating_clasificado'] = df['rating'].apply(lambda x: classify_rating(str(x).lower()))

# Ver el resultado
df[['rating', 'rating_clasificado']].head()
```

```python
def pinta_distribucion_categoricas(df, columnas_categoricas, relativa=False, mostrar_valores=False, giro = 45):
    num_columnas = len(columnas_categoricas)
    num_filas = (num_columnas // 2) + (num_columnas % 2)

    fig, axes = plt.subplots(num_filas, 2, figsize=(15, 5 * num_filas))
    axes = axes.flatten() 

    for i, col in enumerate(columnas_categoricas):
        ax = axes[i]
        if relativa:
            total = df[col].value_counts().sum()
            serie = df[col].value_counts().apply(lambda x: x / total)
            sns.barplot(x=serie.index, y=serie, ax=ax, palette='Reds', hue = serie.index, legend = False)
            ax.set_ylabel('Frecuencia Relativa')
        else:
            serie = df[col].value_counts()
            sns.barplot(x=serie.index, y=serie, ax=ax, palette='Reds', hue = serie.index, legend = False)
            ax.set_ylabel('Frecuencia')

        ax.set_title(f'Distribución de {col}')
        ax.set_xlabel('')
        ax.tick_params(axis='x', rotation=giro)

        if mostrar_valores:
            for p in ax.patches:
                height = p.get_height()
                ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height), 
                            ha='center', va='center', xytext=(0, 9), textcoords='offset points')

    for j in range(i + 1, num_filas * 2):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()
```

```python
# Diccionario basado en palabras clave para los géneros
keyword_genre_mapping = {
    'Reality': 'Reality',
    'Documentary': 'Documental',
    'Comedy': 'Comedia',
    'Drama': 'Drama',
    'Romantic': 'Romántico',
    'Adventure': 'Aventuras',
    'Mystery': 'Misterio',
    'Horror': 'Horror',
    'Family': 'Familia'
}

# Función mejorada para mapear géneros
def classify_genre(row, genre_mapping):
    # Dividir las categorías separadas por coma
    categories = row.split(',') if isinstance(row, str) else []
    
    for category in categories:
        # Limpiar la categoría de espacios y palabras genéricas como 'TV', 'Movies', etc.
        cleaned_category = re.sub(r'\b(TV|Movies|Series|Shows)\b', '', category, flags=re.IGNORECASE).strip()
        
        # Buscar palabras clave en el diccionario
        for keyword, genre in genre_mapping.items():
            if keyword.lower() in cleaned_category.lower():
                return genre  # Retorna el género correspondiente
    
    return 'Otro'  # Si no se encuentra ningún género válido

# Aplicar la función a la columna 'listed_in'
df['genre'] = df['listed_in'].apply(lambda x: classify_genre(x, keyword_genre_mapping))

# Verificar las primeras filas
print(df[['listed_in', 'genre']].head())
```

```python
# Calcular frecuencias absolutas y relativas
genre_counts = df['genre'].value_counts()
genre_relative = df['genre'].value_counts(normalize=True) * 100

# Crear un DataFrame para unir ambas
genre_freq = pd.DataFrame({
    'Genre': genre_counts.index,
    'Absolute Frequency': genre_counts.values,
    'Relative Frequency (%)': genre_relative.values
})

# Gráfico combinado de frecuencias absolutas y relativas
fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Frecuencia Absoluta
sns.barplot(x='Genre',y='Absolute Frequency', data=genre_freq, ax=ax[0], color='darkred')
ax[0].set_title('Frecuencia Absoluta de Géneros')
ax[0].set_ylabel('Frecuencia Absoluta')
ax[0].set_xlabel('Género')

# Frecuencia Relativa
sns.barplot(x='Genre',y='Relative Frequency (%)', data=genre_freq, ax=ax[1], color='black')
ax[1].set_title('Frecuencia Relativa de Géneros')
ax[1].set_ylabel('Frecuencia Relativa (%)')
ax[1].set_xlabel('Género')

plt.tight_layout()
plt.show()
```

Aquí podemos visualizar claramente que en 2015 empiezan a hacerse las series, es algo a tener en cuenta por que su produccion aumenta considerablemte, aqui empezamos a plantearnos,¿Quién empezoó con más titulos de series? ¿Qué país marcó esta tendencia?

```python
def plot_categorical_relationship_fin(df, cat_col1, cat_col2, relative_freq=False, show_values=False, size_group = 5,custom_palette=None):
    # Prepara los datos
    count_data = df.groupby([cat_col1, cat_col2]).size().reset_index(name='count')
    total_counts = df[cat_col1].value_counts()
    
    # Convierte a frecuencias relativas si se solicita
    if relative_freq:
        count_data['count'] = count_data.apply(lambda x: x['count'] / total_counts[x[cat_col1]], axis=1)

    # Si hay más de size_group categorías en cat_col1, las divide en grupos de size_group
    unique_categories = df[cat_col1].unique()
    if len(unique_categories) > size_group:
        num_plots = int(np.ceil(len(unique_categories) / size_group))

        for i in range(num_plots):
            # Selecciona un subconjunto de categorías para cada gráfico
            categories_subset = unique_categories[i * size_group:(i + 1) * size_group]
            data_subset = count_data[count_data[cat_col1].isin(categories_subset)]

            # Crea el gráfico
            plt.figure(figsize=(10, 6))
            ax = sns.barplot(x=cat_col1, y='count', hue=cat_col2, data=data_subset, order=categories_subset,palette=custom_palette)

            # Añade títulos y etiquetas
            plt.title(f'Relación entre {cat_col1} y {cat_col2} - Grupo {i + 1}')
            plt.xlabel(cat_col1)
            plt.ylabel('Frecuencia' if relative_freq else 'Conteo')
            plt.xticks(rotation=45)

            # Mostrar valores en el gráfico
            if show_values:
                for p in ax.patches:
                    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                                ha='center', va='center', fontsize=10, color='black', xytext=(0, size_group),
                                textcoords='offset points')

            # Muestra el gráfico
            plt.show()
    else:
        # Crea el gráfico para menos de size_group categorías
        plt.figure(figsize=(10, 6))
        ax = sns.barplot(x=cat_col1, y='count', hue=cat_col2, data=count_data,palette=custom_palette)

        # Añade títulos y etiquetas
        plt.title(f'Relación entre {cat_col1} y {cat_col2}')
        plt.xlabel(cat_col1)
        plt.ylabel('Frecuencia' if relative_freq else 'Conteo')
        plt.xticks(rotation=45)

        # Mostrar valores en el gráfico
        if show_values:
            for p in ax.patches:
                ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                            ha='center', va='center', fontsize=10, color='black', xytext=(0, size_group),
                            textcoords='offset points')

        # Muestra el gráfico
        plt.show()
```

```python
custom_palette = ['#E50914','#6E6E6E', '#B81D24','#2B2B2B']
```

```python

plot_categorical_relationship_fin(df, "rating_clasificado","type", relative_freq = True, show_values = True,custom_palette=custom_palette)
```

Más alla, cabe destacar que Canada vuelve a ser el país menos poralizado en este aspecto, como ya vimos anteriormente con el consumo por clasificación de edad,pregunta a analizar: ¿Por qué en Canadá los géneros están más equilibrados en comparación con otros países? ¿Influye esto en la diversidad de su audiencia o en la internacionalización del contenido producido allí?

```python
def pareto_principle(series):
    genre_counts = series.value_counts()
    cumulative_percentage = genre_counts.cumsum() / genre_counts.sum() * 100
    return (cumulative_percentage <= 80).sum()

# Aplicar la regla de Pareto
pareto_scores = df_top_countries.groupby('country')['genre'].apply(pareto_principle)
print(pareto_scores)
```

Hemos utilizado el análisis de Pareto para evaluar la distribución de películas por género y país en Netflix, identificando los géneros más representativos según el principio 80/20. En Canadá y Estados Unidos, tres géneros dominan el 80% de la producción, mientras que en Francia, Reino Unido e India solo dos géneros cumplen esta regla, indicando una mayor concentración. Este análisis sugiere que países con menos géneros dominantes, como Francia, tienden a especializarse, mientras que Estados Unidos y Canadá muestran mayor diversidad, aunque Canadá tiene un catálogo más pequeño. Esto podría estar influido por factores demográficos o de mercado.

## ANÁLISIS NUMÉRICO UNIVARIANTE

En este análisis, hemos trabajado con variables numéricas para comprender mejor la duración de películas y series, así como el tiempo entre la realización y el lanzamiento de los títulos. Para ello, fue necesario reestructurar las columnas de duración, separando las películas y series, ya que estas se representan en formatos distintos: minutos para las películas y temporadas para las series. También transformamos la columna de fechas en una variable numérica dividida en años y meses, con el objetivo de extraer patrones temporales que pudieran aportar información valiosa.

```python
# Separar en películas y series
movies = df[df['type'] == 'Movie'].copy()
series = df[df['type'] == 'TV Show'].copy()

# Extraer duración en minutos para películas
movies['duration'] = movies['duration'].str.replace(' min', '').astype(float)

# Extraer número de temporadas para series
series['duration'] = series['duration'].str.replace(' Season', '').str.replace('s', '').astype(float)

# Verificar distribuciones
print(movies['duration'].describe())  # Películas
print(series['duration'].describe())  # Series
```

Con estas variables numéricas nos planteamos, ¿hay una diferencia significativa en la duración de películas frente al número de temporadas en series?  ¿Cuánto tiempo tarda Netflix en agregar contenido desde su año de realización? ¿Las películas recientes tienen una duración diferente a las antiguas? ¿Hay variación en los meses/años en los que Netflix agrega más contenido?

```python
# Convertir a datetime
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extraer año y mes de adición
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Calcular el tiempo transcurrido entre el año de realización y el de adición
df['years_to_add'] = df['year_added'] - df['release_year']

# Verificar resultados
print(df[['release_year', 'year_added', 'years_to_add']].head())
```

```python
# Películas: Histograma de duración
sns.histplot(movies['duration'], bins=30, color='#E50914')
plt.title('Duración de películas en minutos')
plt.xlabel('Duración (min)')
plt.ylabel('Frecuencia')
plt.show()

# Series: Histograma de temporadas
sns.histplot(series['duration'], bins=10, color='#2B2B2B')
plt.title('Número de temporadas en series')
plt.xlabel('Número de temporadas')
plt.ylabel('Frecuencia')
plt.show()

# Tiempo transcurrido entre realización y adición
sns.boxplot(x='type', y='years_to_add', data=df, color='black')
plt.title('Tiempo entre realización y adición')
plt.xlabel('Tipo de producción')
plt.ylabel('Años')
plt.show()
```

En primer lugar, ajustamos la duración de las películas en minutos y de las series en temporadas, representando visualmente ambas distribuciones mediante histogramas. Esto permitió identificar que la duración de las películas sigue una distribución normal con un pico en los 100 minutos, mientras que las series tienden a tener un promedio de dos temporadas. También analizamos el tiempo entre la realización y el lanzamiento o adición de los títulos, utilizando un boxplot. Aunque se observaron outliers en ambos casos, los datos tienden a agruparse cerca de la media, lo que indica cierta consistencia en los tiempos de producción y lanzamiento.

```python
# Resumen estadístico
print("Películas (minutos):")
print(movies['duration'].describe())
print("\nSeries (temporadas):")
print(series['duration'].describe())

# Prueba t para comparar medias (asumiendo normalidad)
from scipy.stats import ttest_ind
t_stat, p_value = ttest_ind(movies['duration'].dropna(), series['duration'].dropna(), equal_var=False)
print(f"Prueba t: Estadístico = {t_stat}, p-valor = {p_value}")
```

La prueba T demuestra que existe una diferencia estadísticamente significativa entre la duración de películas y series, lo que refuerza la necesidad de analizarlas por separado. Además, al profundizar en el tiempo entre realización y edición, las distribuciones permiten observar patrones más detallados que complementan lo que ya se había identificado con los boxplots, ofreciendo una visión más completa de cómo se gestionan los tiempos de producción y lanzamiento en función del tipo de contenido, sin embargo, poco util dentro de nuestro objetivo del análisis.

```python
# Filtrar datos para los cinco países principales
top_countries = ['United States', 'United Kingdom', 'Canada', 'India', 'France']
filtered_df = df[df['country'].isin(top_countries)]

# Agrupar por país y género, y contar los títulos
heatmap_data = filtered_df.groupby(['country', 'genre']).size().unstack(fill_value=0)

# Configurar la figura y colores
plt.figure(figsize=(10, 6))
sns.heatmap(
    heatmap_data,
    cmap=sns.color_palette(['#000000', '#4C0000','#800000','#B22222', '#FF6347', '#FFA07A', '#F5F5F5']),
    cbar_kws={'label': 'Cantidad de títulos'},
    alpha=0.9
)

# Personalización de etiquetas y título en español
plt.title('Cantidad de Títulos por Género y País', fontsize=14, weight='bold')
plt.xlabel('Género', fontsize=12)
plt.ylabel('País', fontsize=12)
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

*	Estados Unidos e India tienen más diversidad, con Drama como el género principal.
