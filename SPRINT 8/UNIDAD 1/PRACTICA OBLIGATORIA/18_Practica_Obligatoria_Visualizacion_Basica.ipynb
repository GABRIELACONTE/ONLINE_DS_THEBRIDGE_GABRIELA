```python
ciudades_mas_pobladas=df.sort_values(by="population_total",ascending=False).head(5)
print(ciudades_mas_pobladas[["city","population_total"]])
```

```python
# Datos originales (supongamos que ya tienes el DataFrame ciudades_mas_pobladas)
ciudades_mas_pobladas = pd.DataFrame({
    'city': ['Los Ángeles', 'San Diego', 'Sacramento', 'Anaheim', 'Santa Ana'],
    'population_total': [3884.307, 1394.928, 513.624, 346.824, 334.227]
})

# Extraer las ciudades y la población total en variables separadas
ciudades = ciudades_mas_pobladas['city'].tolist()  # Lista de nombres de ciudades
poblacion = ciudades_mas_pobladas['population_total'].tolist()  # Lista de poblaciones

# Colores asignados a cada ciudad
colores = ['cornflowerblue', 'gray', '#FF5773', 'olive', 'lightcoral']

# Crear el gráfico
plt.figure(figsize=(5, 5))
plt.bar(ciudades, poblacion, color=colores)

# Agregar los valores encima de las barras
for i, valor in enumerate(poblacion):
    plt.text(i, valor + 50, f"{valor:.1f}", ha='center', fontsize=10)

# Configuración del gráfico
plt.title('Ciudades más pobladas de California', fontsize=10)
plt.xlabel('Ciudades', fontsize=8)
plt.ylabel('Habitantes (miles)', fontsize=8)
plt.ylim(0, 5000)  # Límite del eje Y
plt.xticks(fontsize=9)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar el gráfico
plt.tight_layout()
```

```python
plt.figure(figsize=(5, 5))
plt.hist(df['area_total_km2'], bins=50, color='lightblue', align="left")

# Configurar etiquetas y límites
plt.xlabel('Superficie (km²)', fontsize=18, color='black', fontname='Bradley Hand ITC')
plt.ylabel('Número de ciudades', fontsize=18, color='black', fontname='Bradley Hand ITC')
plt.xlim(-50, 1200)  # Eje X de 0 a 1200
plt.ylim(0, 200)   # Eje Y de 0 a 200
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.margins(x=0)

# Agregar leyenda
plt.legend(['Hist Area'], loc='upper right', fontsize=10, frameon=True, labelcolor='blue');
```

```python
# Calcular el cuartil Q3
Q3 = df['area_water_km2'].quantile(0.75)

# Calcular el límite superior para los outliers
upper_bound = Q3 + 1.5 * (Q3 - df['area_water_km2'].quantile(0.25))

# Filtrar los outliers que están por encima de Q3
outliers = df[df['area_water_km2'] > upper_bound]

# Mostrar los outliers
print(outliers[['city', 'area_water_km2']])
# Ordenar los outliers de mayor a menor área de agua
outliers_sorted = outliers.sort_values(by='area_water_km2', ascending=False)

# Obtener los dos outliers con el área de agua más grande
top_2_outliers = outliers_sorted.head(2)

# Mostrar los dos outliers más grandes
print(top_2_outliers[['population_total', 'area_water_km2']])
```

```python

df=df.dropna(subset=["elevation_m"])
plt.figure(figsize=(5, 5))
boxplot = plt.boxplot(
    df['elevation_m'],
    vert=False,  # Hacer el diagrama horizontal
    patch_artist=True,  # Permitir el relleno de color
    boxprops=dict(color='red', facecolor='red', edgecolor="black"),  # Cuadro rojo
    whiskerprops=dict(color='black'),  # Bigotes negros
    capprops=dict(color='black'),  # Tapas negras
    medianprops=dict(color='black'),  # Mediana negra
    flierprops=dict(marker='o', color='black', markersize=6)  # Outliers negros
)
plt.xlim(-100,2500)
plt.xticks(range(0,2501,500))

# Configurar el gráfico
plt.title('Diagrama de Caja de la Altitud de las Ciudades de California', fontsize=10)
plt.xlabel('Altitud (m)', fontsize=10)
plt.yticks([1], ['1'], fontsize=12)  # Etiqueta del eje Y solo con "1"
# Mostrar el gráfico
plt.show()
```

```python
# Crear la figura con 4 subgráficos (2 filas y 2 columnas)
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
# Primer gráfico: Gráfico de barras de las 5 ciudades más pobladas
axs[0, 0].bar(ciudades, poblacion, color=['#6a5acd', '#808080', '#ff1493', '#556b2f', '#ffcccb'])
axs[0, 0].set_title('Cinco ciudades más pobladas de California')
axs[0, 0].set_xlabel('Ciudades')
axs[0, 0].set_ylabel('Habitantes (Miles)')
axs[0, 0].tick_params(axis='x', labelrotation=45)
# Segundo gráfico: Histograma del área total
axs[0, 1].hist(df['area_total_km2'], bins=50, color='lightblue', align="left")
axs[0, 1].set_title('Histograma del Área Total de Ciudades')
axs[0, 1].set_xlabel('Área Total (km²)')
axs[0, 1].set_ylabel('Número de Ciudades')
# Tercer gráfico: Gráfico de dispersión de área de agua vs población
axs[1, 0].scatter(df['area_water_km2'], df['population_total'], color='green', alpha=0.4)

axs[1, 0].set_title('Población vs Área de Agua')
axs[1, 0].set_xlabel('Área de Agua (km²)')
axs[1, 0].set_ylabel('Población (Miles)')
# Cuarto gráfico: Diagrama de caja de altitud
axs[1, 1].boxplot(df['elevation_m'],
    vert=False,  # Hacer el diagrama horizontal
    patch_artist=True,  # Permitir el relleno de color
    boxprops=dict(color='red', facecolor='red', edgecolor="black"),  # Cuadro rojo
    whiskerprops=dict(color='black'),  # Bigotes negros
    capprops=dict(color='black'),  # Tapas negras
    medianprops=dict(color='black'),  # Mediana negra
    flierprops=dict(marker='o', color='black', markersize=6))  # Outliers negros


axs[1, 1].set_title('Diagrama de Caja - Altitud de Ciudades')
axs[1, 1].set_xlabel('Altitud (m)')
# Ajustar el espacio entre subgráficos para que no se solapen
plt.tight_layout()
# Mostrar la figura
plt.show()
fig.savefig("figura_completa.png",dpi=300)
```
