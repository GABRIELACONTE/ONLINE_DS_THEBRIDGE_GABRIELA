### Ejercicio 2

Reproduce la imagen de la figura usando tres subplots en una columna, utilizando objetos `Axes`. Las dos primeras son las funciones anteriores y la tercera es la tangente (np.tan)

```python
x = np.linspace(0, 10, 1000)  # Valores del eje x de 0 a 10

# Crear la figura y los subgráficos
fig, axs = plt.subplots(3, 1, figsize=(4, 6))  # 3 filas, 1 columna

# Configurar límites comunes
x_limits = (0, 10)




# Gráfico del seno (rojo)
axs[1].plot(x, np.sin(x), color='red')
axs[1].set_xlim(x_limits)
axs[1].set_ylim(-1.5,1.5)
axs[1].set_title("Seno")
# Gráfico del coseno (verde)
axs[0].plot(x, np.cos(x), color='green')
axs[0].set_xlim(x_limits)
axs[0].set_ylim(-1.5,1.5)
axs[0].set_title("Coseno")

# Gráfico de la tangente (azul)
axs[2].plot(x, np.tan(x), color='blue')
axs[2].set_xlim(x_limits)
axs[2].set_ylim(-40,20)
axs[2].set_title("Tangente")
fig
```

### Ejercicio 5: Ajustando el Espaciado de Subplots

Reproduce la siguiente figura con cuatro subplots, ajusta el espaciado entre los subplots para evitar la superosición (tal y como ves en la figura), emplea `subplots_adjust` y los parámetros hspace y wspace.

```python
plt.close("all")
```

```python

x = np.linspace(0, 10, 1000)

# Crear figura con 2 filas y 2 columnas
fig, axs = plt.subplots(2, 2, figsize=(10, 8))  # 2 filas, 2 columnas

# Primera fila, primera columna: Coseno
axs[0, 0].plot(x, np.cos(x))
axs[0, 0].set_title("Coseno")
axs[0, 0].set_xlim(0, 10)
axs[0, 0].set_ylim(-1, 1)
axs[0, 0].set_xticks([0,5,10])

# Primera fila, segunda columna: Seno
axs[0, 1].plot(x, np.sin(x))
axs[0, 1].set_title("Seno")
axs[0, 1].set_xlim(0, 10)
axs[0, 1].set_ylim(-1, 1)
axs[0, 1].set_xticks([0,5,10])

# Segunda fila, primera columna: Tangente
axs[1, 0].plot(x, np.tan(x))
axs[1, 0].set_title("Tangente")
axs[1, 0].set_xlim(0, 10)
axs[1, 0].set_ylim(-40, 20)
axs[1, 0].set_xticks([0,5,10])

# Segunda fila, segunda columna: Seno hiperbólico inverso
axs[1, 1].plot(x, np.arcsinh(x))
axs[1, 1].set_title("Seno hiperbólico inverso")
axs[1, 1].set_xlim(0, 10)
axs[1, 1].set_ylim(0, 3)
axs[1, 1].set_xticks([0,5,10])

# Ajustar el espaciado entre los subplots
fig.subplots_adjust(hspace=0.4, wspace=0.3)
```
