### #1.2

Crea una funcin similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una funcin para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numrica del dataframe
3. Una columna numrica del dataframe diferente a la anterior.
4. Un columna categrica (Opcional, valor por defecto None)
5. Un valor para el tamao de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamao
6. Un sexto argumento "escala", que permita manipular la escala del tamao en el caso de que el quinto argumento sea una cadena. Este argumento multiplicar a los valores de la columna indicada en el parmetro anterior (as si quiero dividir por 10000, este sexto argumento tendr que valer 1/10000)

La funcin debe pintar el diagrama de dispersin de la columna numrica 1, en el eje X, frente a la columna numrica 2, en el eje Y, usando un cdigo de colores en funcin de la columna categrica y con el tamao de los puntos segn el argumento correspondiente. 

Si no se diera una columna categrica como argumento, el color no debe pasarse como parmetro a la funcin de Seaborn o Matplotlib que uses.

EXTRA: Aade un argumento que controle si la funcin pinta una leyenda para colores y tamaos o un colorbar y una leyenda para los tamaos. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesin en vivo)

```python
import matplotlib.pyplot as plt
import seaborn as sns

def scatter_plot_custom(df, x_col, y_col, hue_col=None, size_param=50, scale=1):
    """
    Pinta un diagrama de dispersin con mltiples opciones personalizadas.
    
    Parmetros:
    - df: DataFrame de entrada.
    - x_col: nombre de la columna numrica para el eje X.
    - y_col: nombre de la columna numrica para el eje Y.
    - hue_col: (opcional) nombre de la columna numrica para color (default: None).
    - size_param: valor numrico fijo para el tamao de los puntos o nombre de una columna del DataFrame.
    - scale: factor de escala para el tamao de los puntos si size_param es una columna.
    """
    
    # Definir tamao
    if isinstance(size_param, str):  # Es una columna del DataFrame
        sizes = df[size_param] * scale
    else:  # Es un nmero
        sizes = size_param

    # Crear el grfico
    plt.figure(figsize=(10, 6))
    if hue_col:
        sns.scatterplot(
            data=df, x=x_col, y=y_col, 
            hue=hue_col, 
            palette='viridis',  # Colores verdes/amarillos
            size=sizes, sizes=(20, 200), 
            legend=False
        )
    else:
        sns.scatterplot(
            data=df, x=x_col, y=y_col, 
            size=sizes, sizes=(20, 200), 
            legend=False
        )
    
    plt.title(f"Diagrama de dispersin: {x_col} vs {y_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
dft=pd.read_csv("./data/titanic.csv")
```

```python
dft
```

```python

# Grfico 1: Supervivencia general (frecuencia de vivos/muertos)
sns.countplot(data=dft, x='alive', hue='alive', palette='pastel', legend=False)
plt.title("Supervivencia en el Titanic")
plt.xlabel("Estado")
plt.ylabel("Nmero de personas")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Grfico 2: Supervivencia por gnero
sns.countplot(data=dft, x='sex', hue='alive', palette='pastel')
plt.title("Supervivencia por gnero")
plt.xlabel("Gnero")
plt.ylabel("Nmero de personas")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Grfico 3: Supervivencia por clase
sns.countplot(data=dft, x='pclass', hue='alive', palette='pastel')
plt.title("Supervivencia por clase")
plt.xlabel("Clase")
plt.ylabel("Nmero de personas")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Creamos una columna para identificar a los nios (menores de 18 aos)
dft['is_child'] = dft['age'] < 18

# Grfico 4: Supervivencia por sexo y si es nio
sns.countplot(data=dft, x='is_child', hue='alive', palette='pastel')
plt.title("Supervivencia por sexo y si es nio")
plt.xlabel("Es nio")
plt.ylabel("Nmero de personas")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Grfico 5: Supervivencia segn el tipo de pasaje (Embarked)
sns.countplot(data=dft, x='embarked', hue='alive', palette='pastel')
plt.title("Supervivencia segn tipo de pasaje")
plt.xlabel("Puerto de embarque")
plt.ylabel("Nmero de personas")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
