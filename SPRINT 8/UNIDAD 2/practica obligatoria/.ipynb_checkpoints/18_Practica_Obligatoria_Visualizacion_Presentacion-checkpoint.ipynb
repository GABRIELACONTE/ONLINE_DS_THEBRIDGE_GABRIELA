### #1.2

Crea una función similar a la que se pide en el ejercicio 6 de la segunda tanda de ejercicios del workout. Es decir, una función para pintar diagramas de dispersion que tenga seis argumentos, pero ojo al cambio en el quinto:
1. Un dataframe
2. Una columna numérica del dataframe
3. Una columna numérica del dataframe diferente a la anterior.
4. Un columna categórica (Opcional, valor por defecto None)
5. Un valor para el tamaño de los puntos o una cadena, pero que puede ser una cadena de caracteres, en cuyo caso se debe interpretar como una de las columnas del dataframe cuyos valores deben controlar el tamaño
6. Un sexto argumento "escala", que permita manipular la escala del tamaño en el caso de que el quinto argumento sea una cadena. Este argumento multiplicará a los valores de la columna indicada en el parámetro anterior (así si quiero dividir por 10000, este sexto argumento tendrá que valer 1/10000)

La función debe pintar el diagrama de dispersión de la columna numérica 1, en el eje X, frente a la columna numérica 2, en el eje Y, usando un código de colores en función de la columna categórica y con el tamaño de los puntos según el argumento correspondiente. 

Si no se diera una columna categórica como argumento, el color no debe pasarse como parámetro a la función de Seaborn o Matplotlib que uses.

EXTRA: Añade un argumento que controle si la función pinta una leyenda para colores y tamaños o un colorbar y una leyenda para los tamaños. NO TE PREOCUPES POR EL FORMATO DE LA LEYENDA (lo veremos en la sesión en vivo)

```python
import matplotlib.pyplot as plt
import seaborn as sns

def scatter_plot_custom(df, x_col, y_col, hue_col=None, size_param=50, scale=1):
    """
    Pinta un diagrama de dispersión con múltiples opciones personalizadas.
    
    Parámetros:
    - df: DataFrame de entrada.
    - x_col: nombre de la columna numérica para el eje X.
    - y_col: nombre de la columna numérica para el eje Y.
    - hue_col: (opcional) nombre de la columna numérica para color (default: None).
    - size_param: valor numérico fijo para el tamaño de los puntos o nombre de una columna del DataFrame.
    - scale: factor de escala para el tamaño de los puntos si size_param es una columna.
    """
    
    # Definir tamaño
    if isinstance(size_param, str):  # Es una columna del DataFrame
        sizes = df[size_param] * scale
    else:  # Es un número
        sizes = size_param

    # Crear el gráfico
    plt.figure(figsize=(10, 6))
    if hue_col:
        sns.scatterplot(
            data=df, x=x_col, y=y_col, 
            hue=hue_col, 
            palette='viridis',  # Colores verdes/amarillos
            size=sizes, sizes=(20, 200), 
            legend=False
        )
    else:
        sns.scatterplot(
            data=df, x=x_col, y=y_col, 
            size=sizes, sizes=(20, 200), 
            legend=False
        )
    
    plt.title(f"Diagrama de dispersión: {x_col} vs {y_col}")
    plt.xlabel(x_col)
    plt.ylabel(y_col)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
dft=pd.read_csv("./data/titanic.csv")
```

```python
dft
```

```python

# Gráfico 1: Supervivencia general (frecuencia de vivos/muertos)
sns.countplot(data=dft, x='alive', hue='alive', palette='pastel', legend=False)
plt.title("Supervivencia en el Titanic")
plt.xlabel("Estado")
plt.ylabel("Número de personas")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Gráfico 2: Supervivencia por género
sns.countplot(data=dft, x='sex', hue='alive', palette='pastel')
plt.title("Supervivencia por género")
plt.xlabel("Género")
plt.ylabel("Número de personas")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Gráfico 3: Supervivencia por clase
sns.countplot(data=dft, x='pclass', hue='alive', palette='pastel')
plt.title("Supervivencia por clase")
plt.xlabel("Clase")
plt.ylabel("Número de personas")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Creamos una columna para identificar a los niños (menores de 18 años)
dft['is_child'] = dft['age'] < 18

# Gráfico 4: Supervivencia por sexo y si es niño
sns.countplot(data=dft, x='is_child', hue='alive', palette='pastel')
plt.title("Supervivencia por sexo y si es niño")
plt.xlabel("Es niño")
plt.ylabel("Número de personas")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Gráfico 5: Supervivencia según el tipo de pasaje (Embarked)
sns.countplot(data=dft, x='embarked', hue='alive', palette='pastel')
plt.title("Supervivencia según tipo de pasaje")
plt.xlabel("Puerto de embarque")
plt.ylabel("Número de personas")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
