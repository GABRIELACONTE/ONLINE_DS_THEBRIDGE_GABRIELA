### Ejercicio 2

Crea una funcin a la que puedas pasar como argumentos un dataframe y una lista de columnas categricas y dibuje una matriz de grficas de 2 columnas y tantas filas como necesite para pintar el diagrama de barras de las frecuencias absolutas de las columnas pasadas como argumento. No hace falta que compruebes que las columnas son categricas. Y que ponga como ttulo de cada figura "Frecuencias Absolutas de \<nombre de la columna>\. Utiliza Seaborn para dibujar las grficas y matplotlib para lo que necesites (pero no para dibujar las grficas)

```python


def pinta_distribucion_categoricas(df, columnas_categoricas, relativa=False, mostrar_valores=False):
    num_columnas = len(columnas_categoricas)
    num_filas = (num_columnas // 2) + (num_columnas % 2)

    fig, axes = plt.subplots(num_filas, 2, figsize=(15, 5 * num_filas))
    axes = axes.flatten() 

    for i, col in enumerate(columnas_categoricas):
        ax = axes[i]
        if relativa:
            total = df[col].value_counts().sum()
            serie = df[col].value_counts().apply(lambda x: x / total)
            sns.barplot(x=serie.index, y=serie, ax=ax, palette='viridis', hue = serie.index, legend = False)
            ax.set_ylabel('Frecuencia Relativa')
        else:
            serie = df[col].value_counts()
            sns.barplot(x=serie.index, y=serie, ax=ax, palette='viridis', hue = serie.index, legend = False)
            ax.set_ylabel('Frecuencia')

        ax.set_title(f'Frecuencias absolutas de {col}')
        ax.set_xlabel('')
        ax.tick_params(axis='x', rotation=45)

        if mostrar_valores:
            for p in ax.patches:
                height = p.get_height()
                ax.annotate(f'{height:.2f}', (p.get_x() + p.get_width() / 2., height), 
                            ha='center', va='center', xytext=(0, 9), textcoords='offset points')

    for j in range(i + 1, num_filas * 2):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()
```

### Ejercicio 4

Crea una funcin similar a la del ejercicio 4, pero que esta vez pinte el histograma de una lista de columnas numricas, aade un argumento para poder indicar el nmero de bins. Aade tambin un argumento que permita incluir la funcin de densidad o no en la figura. De nuevo usa Seaborn para crear las figuras. Tambin debe cumplirse el que pinte cada grafica en una matriz de 2 columnas como mucho y todas las filas que necesites.

```python
import math

def plot_histograms(df, numeric_columns, bins=10, kde=False):
    """
    Dibuja una matriz de histogramas para las columnas numricas especificadas.
    
    Args:
    - df: pandas DataFrame.
    - numeric_columns: lista de nombres de columnas numricas a graficar.
    - bins: nmero de bins para los histogramas.
    - kde: booleano para incluir la funcin de densidad (kde) o no.
    """
    # Nmero de columnas en la matriz de grficos
    num_cols = 2
    # Calcular el nmero de filas necesarias
    num_rows = math.ceil(len(numeric_columns) / num_cols)
    
    # Crear figura y ejes
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(8, num_rows * 5))
    axes = axes.flatten()  # Aplanar la matriz de ejes
    
    # Iterar sobre las columnas numricas
    for i, col in enumerate(numeric_columns):
        sns.histplot(df[col], bins=bins, kde=kde, ax=axes[i], color='r')
        axes[i].set_title(f'Histograma de {col}')
        axes[i].set_xlabel(col)
        axes[i].set_ylabel('Frecuencia')
    
    # Eliminar ejes vacos si hay ms subplots que columnas
    for j in range(len(numeric_columns), len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()
```

```python
plot_histograms(dft, numeric_columns=["age","fare"], bins=10, kde=False)
```

### Ejercicio 6

Construye una funcin a la que puedas pasar un dataframe y una serie de columnas numricas (que suponemos son series temporales o secuencias) y que pinte cada columna en una sola grfica (de nuevo en una matriz de 2 columnas como mucho).

```python
def plot_time_series(df, time_series_columns):
    """
    Dibuja una matriz de grficos de lneas para las columnas numricas especificadas.
    
    Args:
    - df: pandas DataFrame.
    - time_series_columns: lista de nombres de columnas numricas a graficar.
    """
    # Nmero de columnas en la matriz de grficos
    num_cols = 2
    # Calcular el nmero de filas necesarias
    num_rows = math.ceil(len(time_series_columns) / num_cols)
    
    # Crear figura y ejes
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, num_rows * 5))
    axes = axes.flatten()  # Aplanar la matriz de ejes
    
    # Iterar sobre las columnas de series temporales
    for i, col in enumerate(time_series_columns):
        axes[i].plot(df.index, df[col], label=col, color='b')
        axes[i].set_title(f'Serie temporal: {col}')
        axes[i].set_xlabel('ndice')
        axes[i].set_ylabel(col)
        axes[i].legend()
    
    # Eliminar ejes vacos si hay ms subplots que columnas
    for j in range(len(time_series_columns), len(axes)):
        fig.delaxes(axes[j])
    
    plt.tight_layout()
    plt.show()
```

### Ejercicio 8

Aade tres argumentos a la funcion del ejercicio 6. El primero debe permitir pintar una sola grfica con todas las lneas de evolucin juntas y aadir una leyenda. El segundo debe permitir colocar una fecha de inicio y el tercero una fecha de fin. Si no se dan valores a estos argumentos la figura debe contemplarlos todos, si se dan la figura slo debe mostrar los valores comprendidos entre esas fechas. Supon que la fecha es el ndice del DataFrame. De nuevo las grficas deben estar pintadas con Seaborn.

```python
def plot_time_series(df, time_series_columns, all_in_one=False, start_date=None, end_date=None):
    """
    Dibuja grficas de lneas para las columnas especificadas.
    Puede graficar todas las columnas juntas o individualmente.
    
    Args:
    - df: pandas DataFrame con las series temporales. Se asume que el ndice es una fecha.
    - columns: lista de nombres de columnas a graficar. Por defecto, ['Open', 'Close', 'Hide', 'Load'].
    - all_in_one: booleano. Si es True, todas las series se pintan juntas con leyenda.
    - start_date: string o datetime. Fecha de inicio para filtrar los datos.
    - end_date: string o datetime. Fecha de fin para filtrar los datos.
    """
    # Filtrar las columnas especficas
    time_series_columns = [col for col in columns if col in df.columns]
    if not time_series_columns:
        raise ValueError("Ninguna de las columnas especificadas est presente en el DataFrame.")
    
    # Filtrar por fechas si se proporcionan
    if start_date or end_date:
        df = df.loc[start_date:end_date]
    
    if all_in_one:
        # Grfica con todas las columnas en una sola figura
        plt.figure(figsize=(12, 6))
        for col in time_series_columns:
            sns.lineplot(x=df.index, y=df[col], label=col)
        plt.title("Evolucin conjunta de las series temporales")
        plt.xlabel("Fecha")
        plt.ylabel("Valor")
        plt.legend()
        plt.tight_layout()
        plt.show()
    else:
        # Grficas individuales en una matriz de subplots
        num_cols = 2
        num_rows = -(-len(time_series_columns) // num_cols)  # Calcular filas (ceil)
        fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, num_rows * 5))
        axes = axes.flatten()  # Aplanar la matriz de ejes
        
        for i, col in enumerate(time_series_columns):
            sns.lineplot(x=df.index, y=df[col], ax=axes[i])
            axes[i].set_title(f'Serie temporal: {col}')
            axes[i].set_xlabel('Fecha')
            axes[i].set_ylabel(col)
        
        # Eliminar ejes vacos si hay ms subplots que columnas
        for j in range(len(time_series_columns), len(axes)):
            fig.delaxes(axes[j])
        
        plt.tight_layout()
        plt.show()
```
